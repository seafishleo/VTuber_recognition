```python
from pytube import YouTube
import os
import glob

# cipher.py Replace (^\w+\W) by (r"^\$*\w+\W")

def download_video(link, save_dir):
    print(f'Download {link} to {save_dir} ...')
    YouTube(link).streams.get_highest_resolution().download(save_dir)
    print('Done\n')


links0 = ['https://www.youtube.com/watch?v=3uLoDspOg0g',
      'https://www.youtube.com/watch?v=zOlxFkUXh1I',
      'https://www.youtube.com/watch?v=WmEnV17WL_g',
      'https://www.youtube.com/watch?v=imbQvCFpZhw',
      'https://www.youtube.com/watch?v=RGACOclbiDM'] # Korone

links1 = ['https://www.youtube.com/watch?v=hmiv9BG-7mk',
      'https://www.youtube.com/watch?v=U0KwtZezl0k',
      'https://www.youtube.com/watch?v=tY4--v4YmdU',
      'https://www.youtube.com/watch?v=lhlcUn_bD20',
      'https://www.youtube.com/watch?v=DrQAsiLDoKQ'] # Okayu

links2 = ['https://www.youtube.com/watch?v=9m6Mmqq_6vY',
      'https://www.youtube.com/watch?v=aCrWIN90whk'] # Mio

links3 = ['https://www.youtube.com/watch?v=OMRoLDdYyLM',
      'https://www.youtube.com/watch?v=zsYfT6T4V6A',
      'https://www.youtube.com/watch?v=2_eoDs7dINQ',
      'https://www.youtube.com/watch?v=PwwgiBtSfeQ',
      'https://www.youtube.com/watch?v=T_k7r9HK-Bg'] # Subaru

links4 = ['https://www.youtube.com/watch?v=sdqjRS2uFec',
      'https://www.youtube.com/watch?v=i4QIHZ3pQ8o',
      'https://www.youtube.com/watch?v=0mYZMXdh3AY',
      'https://www.youtube.com/watch?v=2HghssmG2KQ',
      'https://www.youtube.com/watch?v=k0vwzEqELnM'] # Fubuki

links_all = [links0, links1, links2, links3, links4]

save_dirs = ['/content/drive/MyDrive/VTuber_recognition/video/Korone',
        '/content/drive/MyDrive/VTuber_recognition/video/Okayu',
        '/content/drive/MyDrive/VTuber_recognition/video/Mio',
        '/content/drive/MyDrive/VTuber_recognition/video/Subaru',
        '/content/drive/MyDrive/VTuber_recognition/video/Fubuki']

for (links, dir) in zip(links_all, save_dirs):
    for link in links:
        download_video(link, dir)

    videos = glob.glob(os.path.join(dir, '*.mp4'))
    for (n, video) in enumerate(videos):
        os.rename(video, os.path.join(dir, 'video-{:02}.mp4'.format(n)))
```
