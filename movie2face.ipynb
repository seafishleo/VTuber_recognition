```python
import cv2
import glob
import os
```

```python
def video_to_image(video_paths, out_image_path, num_cut=10):
    img_count = 0
    for video_path in video_paths:
        print(video_path)
        capture = cv2.VideoCapture(video_path)
        frame_count = 0

        while(capture.isOpened()):
            ret, frame = capture.read()
            if ret == False:
                break

            if frame_count % num_cut == 0:         
              img_file_name = os.path.join(out_image_path, '{:05d}.jpg'.format(img_count))
              cv2.imwrite(img_file_name, frame)
              img_count += 1

            frame_count += 1

        capture.release()
    print(f"video_to_image:{img_count}")

def face_detect(out_face_path, img_list):
    xml_path = '/content/drive/MyDrive/VTuber_recognition/lbpcascade_animeface.xml'
    classifier = cv2.CascadeClassifier(xml_path)

    img_count = 0
    for img_path in img_list:
      org_img = cv2.imread(img_path, cv2.IMREAD_COLOR)
      gray_img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
      face_points = classifier.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=2, minSize=(30,30))

      for points in face_points:
          x, y, width, height = points
          dst_img = org_img[y:y+height, x:x+width]
          face_img = cv2.resize(dst_img, (128,128))
          new_img_name = os.path.join(out_face_path, '{:05d}.jpg'.format(img_count))
          cv2.imwrite(new_img_name, face_img)
          img_count += 1
    print(f"face detected:{img_count}")

VTubers = ['Korone', 'Okayu', 'Mio', 'Subaru', 'Fubuki']
for VTuber in VTubers:
    print(VTuber)
    video_dir = f'/content/drive/MyDrive/VTuber_recognition/video/{VTuber}'
    video_paths = glob.glob(os.path.join(video_dir, '*.mp4'))

    image_dir = f'/content/drive/MyDrive/VTuber_recognition/image/{VTuber}'
    images = glob.glob(os.path.join(image_dir, '*.jpg'))

    out_image_path = f'/content/drive/MyDrive/VTuber_recognition/image/{VTuber}/'
    out_face_path = f'/content/drive/MyDrive/VTuber_recognition/face/{VTuber}/'

    print('Video to image ...')
    video_to_image(video_paths, out_image_path, num_cut=10)  # save image every 10 frame

    print('Save %s faces ...' %(VTuber))
    face_detect(out_face_path, images)  # save face images
```

```python
def video_to_image(video_paths, out_image_path, num_cut=10):
    img_count = 0
    for video_path in video_paths:
        print(video_path)
        capture = cv2.VideoCapture(video_path)
        frame_count = 0

        while(capture.isOpened()):
            ret, frame = capture.read()
            if ret == False:
                break

            if frame_count % num_cut == 0:         
              img_file_name = os.path.join(out_image_path, '{:05d}.jpg'.format(img_count))
              cv2.imwrite(img_file_name, frame)
              img_count += 1

            frame_count += 1

        capture.release()
    print(f"video_to_image:{img_count}")

def face_detect(out_face_path, img_list):
    xml_path = '/content/drive/MyDrive/VTuber_recognition/lbpcascade_animeface.xml'
    classifier = cv2.CascadeClassifier(xml_path)

    img_count = 0
    for img_path in img_list:
      image = cv2.imread(img_path, cv2.IMREAD_COLOR)
      gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
      gray = cv2.equalizeHist(gray)
      face_points = classifier.detectMultiScale(gray, scaleFactor = 1.1, minNeighbors = 5, minSize = (24, 24))

      for points in face_points:
        x, y, w, h = points
        face_img = cv2.rectangle(image, (x, y), (x + w, y + h), (0, 0, 255), 2)
        new_img_name = os.path.join(out_face_path, '{:05d}.jpg'.format(img_count))
        cv2.imwrite(new_img_name, face_img)
        img_count += 1
    print(f"face detected:{img_count}")

VTubers = ['Korone', 'Okayu', 'Mio', 'Subaru', 'Fubuki']
for VTuber in VTubers:
    print(VTuber)
    video_dir = f'/content/drive/MyDrive/VTuber_recognition/video/{VTuber}'
    video_paths = glob.glob(os.path.join(video_dir, '*.mp4'))

    image_dir = f'/content/drive/MyDrive/VTuber_recognition/image/{VTuber}'
    images = glob.glob(os.path.join(image_dir, '*.jpg'))

    out_image_path = f'/content/drive/MyDrive/VTuber_recognition/image/{VTuber}/'
    out_face_path = f'/content/drive/MyDrive/VTuber_recognition/face/{VTuber}/'

    print('Video to image ...')
    video_to_image(video_paths, out_image_path, num_cut=10)  # save image every 10 frame

    print('Save %s faces ...' %(VTuber))
    face_detect(out_face_path, images)  # save face images
```
